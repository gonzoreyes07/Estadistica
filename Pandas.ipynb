```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Leemos el archivo con la librer√≠a pandas
df = pd.read_csv("Alumnos.csv", encoding="latin-1")

# Unimos nombre y apellido en una sola columna
df["Nombre Completo"] = df.iloc[:, 0] + " " + df.iloc[:, 1]

# Agregar columnas con datos aleatorios
np.random.seed(42)
df["Edad"] = np.random.randint(18, 26, size=len(df))
df["Calificaci√≥n"] = np.random.normal(7.5, 1.0, size=len(df)).clip(0, 10)  # Calificaci√≥n entre 0 y 10
df["Asistencia (%)"] = np.random.uniform(70, 100, size=len(df))

# Mostrar el DataFrame
print(df.head())

# 1. Histograma de distribuci√≥n de edades
plt.figure(figsize=(8, 5))
plt.hist(df["Edad"], bins=8, color="green", edgecolor="black")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.title("Distribuci√≥n de edades de los alumnos")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

# 2. Gr√°fico de barras de asistencia
plt.figure(figsize=(10, 6))
plt.bar(df["Nombre Completo"], df["Asistencia (%)"], color="lightcoral")
plt.xlabel("Alumnos")
plt.ylabel("Asistencia (%)")
plt.title("Porcentaje de Asistencia de los Alumnos")
plt.xticks(rotation=90)
plt.ylim(50, 100)  # Ajuste del rango de asistencia
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()

# 3. Diagrama de dispersi√≥n: Edad vs. Asistencia
plt.figure(figsize=(8, 5))
plt.scatter(df["Edad"], df["Asistencia (%)"], color="green", alpha=0.6)
plt.xlabel("Edad")
plt.ylabel("Asistencia (%)")
plt.title("Relaci√≥n entre Edad y Asistencia")
plt.grid(True)
plt.show()

# 4. Gr√°fico de barras de Calificaciones
plt.figure(figsize=(10, 6))
plt.bar(df["Nombre Completo"], df["Calificaci√≥n"], color="blue")
plt.xlabel("Alumnos")
plt.ylabel("Calificaci√≥n")
plt.title("Calificaciones de los Alumnos")
plt.xticks(rotation=90)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()

# 5. Estad√≠sticas generales del DataFrame
print(df.describe())
```
