```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
```

```python
n=50
p=0.7
muestras=10000
np.random.seed(42)
datos=np.random.binomial(n,p,size=muestras)
```

```python
plt.hist(datos,bins=range(n+2), density=True, alpha=0.6, edgecolor="black")
plt.title("Simulación de una distribuñción Binomial")
plt.xlabel("Número de Exitos")
plt.ylabel("Probabilidad")
plt.show()
```

```python
# Esperanza empírica de la simulación
esperanza_empirica=np.mean(datos)
print("La esperanza empírica de la simulación es:", esperanza_empirica)

# Esperanza teórica
esperanza_teorica=n*p
print("La esperanza teórica de la simulación es:", esperanza_teorica)

# Varianza empírica
varianza_empirica=np.var(datos)
print("La varianza empírica de la simulación es:", varianza_empirica)

# Varianza teórica
varianza_teorica=n*p*(1-p)
print("La varianza teórica de la simulación es:", varianza_teorica)
```
